```python
# Check for missing values in each column
facts_data.isnull().any()
```

```python
# Get the column names with no values
empty_columns = facts_data.columns[facts_data.isnull().all()]
print("Empty columns:")
print(empty_columns)
```

```python
facts_data.drop(columns = empty_columns, inplace = True)
```

```python
# Get columns with mixed types
mixed_type_columns = []

# Iterate over each column
for column in facts_data.columns:
    unique_types = facts_data[column].apply(type).unique()
    if len(unique_types) > 1:
        mixed_type_columns.append(column)

print("Columns with mixed types:")
print(mixed_type_columns)
```

```python
facts_data.drop_duplicates(subset=['code'],inplace=True)
```
